#1. Análisis de facturación
En este primer análisis, debes calcular el ingreso total de cada tienda. Sumando los valores de la columna Precio de cada conjunto de datos de la tienda para estimar los ingresos.



Conversion de DataFrame a directorios para poder trabajar con phyton directamente

```python
datos_tienda1 = tienda.to_dict(orient='records')
datos_tienda2 = tienda2.to_dict(orient='records')
datos_tienda3 = tienda3.to_dict(orient='records')
datos_tienda4 = tienda4.to_dict(orient='records')
```

Creacion de una función para poder acortar las lineas de codigo.

```python
def calcular_ingresos(datos):
    total = sum(ventas['Precio'] for ventas in datos)
    return round(total, 2)
```

```python
ingresos_tienda1 = calcular_ingresos(datos_tienda1)
print(f"Ingresos de la tienda 1: ${ingresos_tienda1:,.2f}")

ingresos_tienda2 = calcular_ingresos(datos_tienda2)
print(f"Ingresos de la tienda 2: ${ingresos_tienda2:,.2f}")

ingresos_tienda3 = calcular_ingresos(datos_tienda3)
print(f"Ingresos de la tienda 3: ${ingresos_tienda3:,.2f}")

ingresos_tienda4 = calcular_ingresos(datos_tienda4)
print(f"Ingresos de la tienda 4: ${ingresos_tienda4:,.2f}")
```

# 2. Ventas por categoría
Calcular la cantidad de productos vendidos por categoría en cada tienda. La idea es agrupar los datos por categoría y contar el número de ventas de cada tipo, mostrando las categorías más populares de cada tienda.

```python
def ventas_x_categoria(datos_tienda):
    categorias = {}
    for venta in datos_tienda:
        categoria = venta['Categoría del Producto']
        if categoria in categorias:
            categorias[categoria] += 1
        else:
            categorias[categoria] = 1
    categorias_ordenadas = dict(sorted(categorias.items()))
    return categorias_ordenadas
```

```python
ventas_x_categoria_tienda1 = ventas_x_categoria(datos_tienda1)
#print(f"Ventas por categoría de la tienda 1: {ventas_x_categoria_tienda1}")
print("Ventas por categoría de la tienda 1:")
print("\n")

for categoria, cantidad in ventas_x_categoria_tienda1.items():
    print(f"{categoria}: {cantidad}")

print("\n")
print("Ventas por categoría de la tienda 2:")
print("\n")

ventas_x_categoria_tienda2 = ventas_x_categoria(datos_tienda2)

for categoria, cantidad in ventas_x_categoria_tienda2.items():
    print(f"{categoria}: {cantidad}")

print("\n")
print("Ventas por categoría de la tienda 3:")
print("\n")

ventas_x_categoria_tienda3 = ventas_x_categoria(datos_tienda3)

for categoria, cantidad in ventas_x_categoria_tienda3.items():
    print(f"{categoria}: {cantidad}")

print("\n")
print("Ventas por categoría de la tienda 4:")
print("\n")

ventas_x_categoria_tienda4 = ventas_x_categoria(datos_tienda4)

for categoria, cantidad in ventas_x_categoria_tienda4.items():
    print(f"{categoria}: {cantidad}")
print("\n")

```

# 3. Calificación promedio de la tienda
En este paso, debes calcular las calificaciones promedio de los clientes para cada tienda. El objetivo es conocer la satisfacción del cliente con los productos vendidos.




Validamos el sistema de calificación para asegurarnos que esté en la escala de 1 a 5

```python
def explorar_calificaciones(datos_tienda):
    calificaciones = set()
    for venta in datos_tienda:
        calificacion = venta['Calificación']
        calificaciones.add(calificacion)
    return sorted(calificaciones)
```

```python
tipo_d_clasificacion = explorar_calificaciones(datos_tienda1)
print("Calificaciones encontradas en la tienda 1:",tipo_d_clasificacion)

tipo_d_clasificacion = explorar_calificaciones(datos_tienda2)
print("Calificaciones encontradas en la tienda 2:",tipo_d_clasificacion)

tipo_d_clasificacion = explorar_calificaciones(datos_tienda3)
print("Calificaciones encontradas en la tienda 3:",tipo_d_clasificacion)

tipo_d_clasificacion = explorar_calificaciones(datos_tienda4)
print("Calificaciones encontradas en la tienda 4:",tipo_d_clasificacion)
```

El promedio

```python
def clasificacion_prom (datos_tienda):
  total=0
  cantidad=0
  for venta in datos_tienda:
    total += venta['Calificación']
    cantidad += 1
  if cantidad == 0:
    return 0
  return total / cantidad
```

```python
promedio1 = clasificacion_prom(datos_tienda1)
promedio2 = clasificacion_prom(datos_tienda2)
promedio3 = clasificacion_prom(datos_tienda3)
promedio4 = clasificacion_prom(datos_tienda4)

print(f"EL promedio calificación Tienda 1 es de : {promedio1:.2f} estrellas")
print(f"EL promedio calificación Tienda 2 es de : {promedio2:.2f} estrellas")
print(f"EL promedio calificación Tienda 3 es de : {promedio3:.2f} estrellas")
print(f"EL promedio calificación Tienda 4 es de : {promedio4:.2f} estrellas")
```

# 4. Productos más y menos vendidos
En este paso, debes identificar qué productos fueron los más vendidos y los menos vendidos en cada tienda. Visualiza los resultados para que quede claro qué productos destacaron en ventas en cada tienda.

```python
def mas_menos_vendidos(ventas_x_categoria,numero_tienda):
   mas_vendidos= max(ventas_x_categoria.items(), key=lambda x: x[1])
   menos_vendidos = min(ventas_x_categoria.items(), key=lambda x: x[1])
   print(f"Tienda {numero_tienda}")
   print(f"Producto más vendido: {mas_vendidos[0]} con {mas_vendidos[1]} ventas")
   print(f"Producto menos vendido: {menos_vendidos[0]} con {menos_vendidos[1]} ventas")
   print("\n")
```

```python
mas_menos_vendidos(ventas_x_categoria_tienda1,1)
mas_menos_vendidos(ventas_x_categoria_tienda2,2)
mas_menos_vendidos(ventas_x_categoria_tienda3,3)
mas_menos_vendidos(ventas_x_categoria_tienda4,4)
```

# 5. Envío promedio por tienda
En este paso, debes calcular el costo de envío promedio para cada tienda. El objetivo es comprender cuánto se gasta, en promedio, en el envío de cada tienda.

```python
def prom_envio(tienda,numero_tienda):
    total_envio = 0
    cantidad_envios = 0
    promedio=0
    for venta in tienda:
        if venta['Costo de envío'] > 0:
            total_envio += venta['Costo de envío']
            cantidad_envios += 1
    if cantidad_envios > 0:
        promedio = total_envio / cantidad_envios
        print(f" Tienda {numero_tienda} - El costo promedio de envío para esta tienda es de: ${promedio:,.2f}")
    else:
        print(f"Tienda {numero_tienda} - No hay datos de envío con costo.")
    return promedio
```

```python
prom_envio_1 = prom_envio(datos_tienda1,1)
prom_envio_2 = prom_envio(datos_tienda2,2)
prom_envio_3 = prom_envio(datos_tienda2,2)
prom_envio_4 = prom_envio(datos_tienda3,3)
```
